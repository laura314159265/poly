```python
import numpy as np
import matplotlib.pyplot as plt

# Dades en unitats del SI
G= 6.67430e-11
m_sol= 1.98847e30
m_terra= 5.97219e24
m_lluna= 7.349e22
m_mart= 6.4171e23
m_jupiter= 18.9819e26
m_venus= 48.685e23

# Condicions inicials (en cartesianes)
# L'origen s el baricentre del sistema solar
# Format: [x,y,z,vx,vy,vz]
# Inici a les 13:59 del 21 de Juny del 2025 (solstici d'estiu)
# Aix s per el temps central Europeu, per tant hi ha un lleuger error
# Ja que no est a Bellaterra/Barcelona exactament
# Posicions en kilometres i velocitats en km/s

lluna= np.array([-1.223926839954494E+05, -1.526112945830738E+08, 5.512426765490323E+04, 2.865929048320844E+01, 7.673984990113614E-01, 6.367954754048094E-02])
terra= np.array([-4.133776082794621E+05, -1.528302631740358E+08, 3.242787546376139E+04, 2.932642969193051E+01, -7.378251008835289E-02, -1.172300942425906E-03])
sol= np.array([-6.709778526787314E+05, -8.035450223548296E+05, 2.366174814126350E+04, 1.263745636318025E-02, -3.212119035252178E-03, -2.265926685896345E-04])
mart= np.array([-2.457240245390437E+08, -2.253913894859833E+07, 5.577461384361338E+06, 3.057555929366051E+00, -2.206677199548648E+01, -5.372712135532627E-01])
jupiter= np.array([-3.766214907360811E+07, 7.674900530284423E+08, -2.340181864259541E+06, -1.320277350206942E+01, -1.821231426227618E-02, 2.954873217258936E-01])
venus= np.array([9.000848903739944E+07, -6.083188193144653E+07, -6.033255262740046E+06, 1.912973897447851E+01, 2.905777575965793E+01, -7.040412673920500E-01])

# Normalitzaci
t0= 86400.0            # Temps caracterstic= 1 dia
m0= m_sol              # Massa caracterstica= massa del sol
r0= (G * t0**2 * m0)**(1/3)  # Radi caracterstic a partir de les dues constants anteriors

def normalitza(cos): # Primer passem a m i m/s, desprs normalitzem
  pos= cos[:3] * 1e3 / r0
  vel= cos[3:] * 1e3 * (t0 / r0)
  return pos, vel

masses= np.array([m_sol/m0, m_terra/m0, m_lluna/m0, m_mart/m0, m_jupiter/m0, m_venus/m0])
pos_sol, vel_sol= normalitza(sol)
pos_terra, vel_terra= normalitza(terra)
pos_lluna, vel_lluna= normalitza(lluna)
pos_mart, vel_mart= normalitza(mart)
pos_jupiter, vel_jupiter= normalitza(jupiter)
pos_venus, vel_venus= normalitza(venus)

posicions= np.array([pos_sol, pos_terra, pos_lluna, pos_mart, pos_jupiter, pos_venus])
velocitats= np.array([vel_sol, vel_terra, vel_lluna, vel_mart, vel_jupiter, vel_venus])

def acceleracions(pos, masses): # pos s matriu N files i 3 columnes
    N= len(masses) # Nombre de planetes
    acc= np.zeros_like(pos) # Crea un array amb la mateixa forma que pos per ple de zeros
    for i in range(N):
        for j in range(N):
            if i != j:  # Si i  j, perqu un cos no s'atrau a ell mateix
                diff= pos[j] - pos[i]
                d= np.linalg.norm(diff) # Calcula la norma del vector diferncia de posicions
                acc[i]= acc[i] + masses[j] * diff / d**3
    return acc

def euler(pos, vel, masses, h):
    acc_euler= acceleracions(pos, masses)
    vel_nou= vel + h * acc_euler
    pos_nou= pos + h * vel
    return pos_nou, vel_nou

def rk4(pos, vel, masses, h):
    K1= vel
    L1= acceleracions(pos, masses)
    K2= vel + 0.5*h*L1
    L2= acceleracions(pos + 0.5*h*K1, masses)
    K3= vel + 0.5*h*L2
    L3= acceleracions(pos + 0.5*h*K2, masses)
    K4= vel + h*L3
    L4= acceleracions(pos + h*K3, masses)
    pos_nou= pos + (h/6)*(K1 + 2*K2 + 2*K3 + K4)
    vel_nou= vel + (h/6)*(L1 + 2*L2 + 2*L3 + L4)
    return pos_nou, vel_nou

# Apliquem aquestes funcions a una simulaci de duraci 1 any terrestre
dies= 365
h= 1/24
passos= int(dies/h)
trajectoria_euler= np.zeros((passos+1, 6, 3))
trajectoria_rk4  = np.zeros((passos+1, 6, 3))
pos_euler, vel_euler= posicions.copy(), velocitats.copy()
pos_rk4, vel_rk4= posicions.copy(), velocitats.copy()
trajectoria_euler[0] = pos_euler
trajectoria_rk4[0]  = pos_rk4

for k in range(passos):
    pos_euler, vel_euler= euler(pos_euler, vel_euler, masses, h)
    pos_rk4, vel_rk4= rk4(pos_rk4, vel_rk4, masses, h)
    trajectoria_euler[k+1]= pos_euler
    trajectoria_rk4[k+1]  = pos_rk4

# GRFIC COMPARATIU
fig= plt.figure(figsize=(9,9))
ax= fig.add_subplot(111, projection="3d")
ax.scatter(trajectoria_rk4[0,0,0], trajectoria_rk4[0,0,1], trajectoria_rk4[0,0,2], color="orange", s=100, label="Sol")

ax.plot(trajectoria_euler[:,1,0], trajectoria_euler[:,1,1], trajectoria_euler[:,1,2], label="Terra (Euler)", color="limegreen", linestyle="--", zorder=2)
ax.plot(trajectoria_rk4[:,1,0], trajectoria_rk4[:,1,1], trajectoria_rk4[:,1,2], label="Terra (RK4)", color="green", zorder=2)
ax.plot(trajectoria_euler[:,2,0], trajectoria_euler[:,2,1], trajectoria_euler[:,2,2], label="Lluna (Euler)", color="purple", linestyle="--", zorder=3)
ax.plot(trajectoria_rk4[:,2,0], trajectoria_rk4[:,2,1], trajectoria_rk4[:,2,2], label="Lluna (RK4)", color="darkorchid", zorder=3)

ax.plot(trajectoria_euler[:,3,0], trajectoria_euler[:,3,1], trajectoria_euler[:,3,2], label="Mart (Euler)", color="red", linestyle="--", zorder=4)
ax.plot(trajectoria_rk4[:,3,0], trajectoria_rk4[:,3,1], trajectoria_rk4[:,3,2],label="Mart (RK4)", color="maroon", zorder=4)

ax.plot(trajectoria_euler[:,4,0], trajectoria_euler[:,4,1], trajectoria_euler[:,4,2], label="Jupiter (Euler)", color="gold", linestyle="--", zorder=5)
ax.plot(trajectoria_rk4[:,4,0], trajectoria_rk4[:,4,1], trajectoria_rk4[:,4,2], label="Jupiter (RK4)", color="orange", zorder=5)

ax.plot(trajectoria_euler[:,5,0], trajectoria_euler[:,5,1], trajectoria_euler[:,5,2], label="Venus (Euler)", color="darkblue", linestyle="--", zorder=1)
ax.plot(trajectoria_rk4[:,5,0], trajectoria_rk4[:,5,1], trajectoria_rk4[:,5,2], label="Venus (RK4)", color="dodgerblue", zorder=1)


ax.set_xlabel("x (adimensional)")
ax.set_ylabel("y (adimensional)")
ax.set_zlabel("z (adimensional)")
ax.set_title("rbites 3D: comparaci")

ax.view_init(90, 0)

ax.legend()
plt.show()

#Triga 14s a plotar
```
